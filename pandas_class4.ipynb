# 🐼 Pandas - Class 4: Data Selection & Filtering
Welcome to **Class 4** of our Pandas series. Today we’ll explore techniques to select and filter data in a DataFrame.

## 1. Boolean Indexing & Conditional Selection
- Use comparison operators to build conditions (e.g., `df['col'] > 50`).
- Pass the condition inside `df[...]` to get matching rows.
- Combine multiple conditions with `&` (AND), `|` (OR), and `~` (NOT).
- Parentheses are required when combining conditions.

```python
import pandas as pd

a = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'Score': [85, 90, 78, 92, 88]
}

b = pd.DataFrame(a)
b
```

```python
#1. Select single column

b[b['Age'] > 27]
```

```python
#2. Selecting single column score > 85

b[b['Score'] > 85]
```

```python
#3. Multiple Conditions (AND): Age > 27 AND Score > 90

b[(b['Age'] > 27) & (b['Score'] > 90)]
```

```python
#4. Multiple Conditions (OR): Age >= 40 or Score > 90

b[(b['Age'] >= 40) | (b['Score'] > 90)]
```

```python
#5. Multiple Conditions (Not): Score not > 85

b[~(b['Score'] > 85)]
```

```python
#6. Age > 22 and Score > 80

b[(b['Age'] > 22) & (b['Score'] > 80)]
```

## 2. Filtering Rows with Multiple Conditions
- Combine two or more conditions to refine selection.
- Remember to enclose each condition in parentheses.

```python
#1. Age > 30 and Score > 85

b[(b['Age'] > 30) & (b['Score'] > 85)]
```

```python
#2. Age < 35 or Score >= 90

b[(b['Age']> 35) | (b['Score'] > 90)]
```

```python
#3. Age between 25 and 35 (both inclusive) using not operator

b[~(b['Age'] <= 25) & ~(b['Age'] >= 35)]
```

## 3. Using `query()` for Cleaner Filtering
- `query()` allows filtering rows with a string expression.
- It's concise and readable for complex conditions.
- Variables can be passed using the `@` symbol.

```python
#1. Simple Query : Age > 25

print(b.query("Age > 25"))
```

```python
#2. Mutliple Conditions using query

print(b.query("Age > 25 & Score > 90"))
```

```python
print(b.query("Age > 30 | Score > 90"))
```

```python
b
```

```python
print(b.query("~(Age < 25) & ~(Age > 35)"))
```

```python
#3. Threshold (@)

score = 90 #Variable
b.query("Score > @score") # Score > 90
```

## 4. String Operations on Columns
- Use the `.str` accessor for vectorized string methods.
- Common methods: `contains`, `replace`, `startswith`, `endswith`, `lower`, `upper`.
- Useful for cleaning or filtering text data.

```python
#1. Convert name to lowercase()

print(b['Name'].str.lower()) # str -> converts into string
```

```python
#2. Convert name to uppercase()

print(b['Name'].str.upper())
```

```python
#3. title() -> Capitalize first letter

print(b['Name'].str.title())
```

```python
#3. Contains()

print(b['Name'].str.contains('Bob'))
```

```python
#4. Replace()

print(b['Name'].str.replace('Bob', 'Bobby'))
```

```python
#5. startswith()

print(b['Name'])
print(b['Name'].str.startswith('a')) # Gives False
print(b['Name'].str.startswith('A'))
```

```python
#6. endswith()

print(b['Name'])
print(b['Name'].str.endswith('e'))
```
