
# Q10. Concatenation and Stacking

## Problem Statement
Perform concatenate and vstack on two arrays.

## Sample Input
```python
Use NumPy arrays
```

## Sample Output
```python
Expected Output
```


````python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a + b)
print(np.hstack([a,b]))
print(np.vstack([a,b]))
````

````python
import numpy as np

a = np.array([3, 1, 4, 1, 5])
sorted_a = np.sort(a)      # [1, 1, 3, 4, 5]
a.sort()                   # a is now [1, 1, 3, 4, 5]
````

````python
b = np.array([[3, 1, 2],
              [9, 7, 8]])

np.sort(b, axis=0)
# array([[3, 1, 2],
#        [9, 7, 8]])  # columns sorted independently

np.sort(b, axis=1)
# array([[1, 2, 3],
#        [7, 8, 9]])  # rows sorted independently
````

````python
b.sort(axis=1) # sorts rows of b in place
b
````

````python
a = np.array([30, 10, 20])
idx = np.argsort(a)    # [1, 2, 0]
sorted_a = a[idx]      # [10, 20, 30]
````

````python
b = np.array([[3, 1, 2],
              [9, 7, 8]])

idx_rows = np.argsort(b, axis=1)  # indices to sort each row
idx_rows
````

````python
data = np.array([[101, 50],
                 [102, 20],
                 [103, 80]])

# sort by column 1 (scores)
order = np.argsort(data[:, 1])
sorted_data = data[order]
sorted_data
````

````python
order_new = np.argsort(data)
order_new
````

````python
order_col = np.sort(data, axis=0)
order_col
````

````python
order_row = np.sort(data, axis=1)
order_row
````
