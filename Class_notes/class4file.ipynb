##Fancy Indexing
you can Pick elements from specific positions using a list/array of indices.

````python
import numpy as np

arr = np.arange(0,70,10)
arr
````

````python
arr[[0,2,4]]
````

````python
# Pick elements at position 1,0,5
indices = np.array([1,0,5])
arr[indices]
````

You can do this fancy indexing in 2D Matrix aswell

````python
import numpy as np
matrix = np.arange(1,13).reshape(3,4)
matrix
````

````python
print(matrix[[0,1],[1,2]])
````

````python
print(matrix[[1,2,1],[0,2,3]])
````

##Conditional Selection
Find values based on conditions or replace them.

````python
arr = np.array([5,7,11,18,30])

# We will get the indices only
# which are greater than 10
np.where(arr > 10)
````

````python
np.where(arr > 10, arr, 0) # replace in arr with 0
````

In your line:

```python
np.where(arr > 10, arr, 0)

```

- ‎`condition` is ‎`arr > 10`

- ‎`x` is ‎`arr`

- ‎`y` is ‎`0`

The rule of ‎`np.where(condition, x, y)` is:

for each element → if ‎`condition` is True → take from ‎`x`, else → take from ‎`y`.

Your ‎`arr` is:

```python
arr = np.array([5, 7, 11, 18, 30])

```

Let’s see the condition ‎`arr > 10`:

- 5  > 10 → False → take from ‎`0` → 0

- 7  > 10 → False → take from ‎`0` → 0

- 11 > 10 → True  → take from ‎`arr` → 11

- 18 > 10 → True  → 18

- 30 > 10 → True  → 30



````python
arr2 = np.array([23,25,30,50,100,75,80])

np.where(arr2 > 55, arr2, 0)
````

##Sorting & Ordering
np.sort() → returns sorted array

np.argsort() → returns indices of sorted order

````python
arr = np.array([2, 10, 56, 1, 54])
np.sort(arr)
````

````python
np.argsort(arr)
````
