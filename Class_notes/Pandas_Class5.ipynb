# 🐼 Pandas - Class 5: Sorting & Basic Statistics
Welcome to **Class 5** of our Pandas series. Today we’ll learn how to sort data and calculate basic statistics.

## 1. Sorting Data
- Use `sort_values(by='col')` to sort by column values.
- Use `ascending=False` for descending order.
- Sort by multiple columns by passing a list.
- `sort_index()` sorts by row or column index.

```python
import pandas as pd

data = pd.DataFrame({
    "Product": ['Pen', 'Notebook', 'Pencil', 'Marker', 'Eraser', 'Sharpener'],
    'Price': [10, 50, 5, 30, 8 ,12],
    'Stock': [100, 60, 200, 80, 150, 120],
    'Rating': [4.5, 4.7, 4.2, 4.8, 4.3, 4.4]
})

df = data
df
```

```python
#1. sort by a single column -> Price (ascending)

df_sorted = df.sort_values(by='Price')
df_sorted
```

```python
df_sorted_desc = df.sort_values(by='Price', ascending=False)
df_sorted_desc
```

```python
#2. Sort by a single column -> Rating (descending)

df_sorted_rating = df.sort_values(by='Rating', ascending=False)
df_sorted_rating
```

```python
#3. Sorting multiple columns -> Price and Stock

df_sorted_price_stock = df.sort_values(by=['Price', 'Stock'])
df_sorted_price_stock
```

```python
# sorting all the columns

df_sorted_all_columns = df.sort_values(by=['Price', 'Stock', 'Rating'], ascending=[True, False, False])
df_sorted_all_columns
```

```python
#4. sort_index() -> in descending order.

df_sorted_index = df.sort_index(ascending=False)
df_sorted_index
```

## 3. Counting Values
- `value_counts()` shows the frequency of unique values in a Series.

## 4. Correlation & Covariance
- `corr()` computes correlation between numeric columns.
- `cov()` computes covariance.
- Correlation values range from -1 (negative) to +1 (positive).

```python
#1. Correlation between all numeric columns

# +ve correlation -> one increase and other increases
# -ve correlation -> one decrease and other increases and vice-versa

df.corr(numeric_only= True)
```

Covariance is a statistical measure that tells you how two numerical variables vary with respect to each other. Think of having two columns of numbers, X and Y (for example: hours studied and exam score). Covariance looks at each pair of values and checks:

- When X is above its mean, is Y also usually above its mean?

- When X is below its mean, is Y also usually below its mean?

If yes, the covariance is positive, meaning the variables tend to move together.

If when X is above its mean, Y is usually below its mean (and vice versa), the covariance is negative, meaning the variables tend to move in opposite directions.



```python
#2. Covariance between all numeric columns

df.cov(numeric_only= True)
```

```python
#3. Correlation between 2 columns Price and Stock

df['Price'].corr(df['Stock'])
```
