Welcome to **Class 7** of our Pandas series. Today we’ll learn how to combine multiple DataFrames using different techniques.

## 1. Using `concat()` for Stacking Data
- `pd.concat()` combines DataFrames vertically (rows) or horizontally (columns).
- Use `axis=0` for vertical stacking (default), `axis=1` for horizontal.
- `ignore_index=True` can reset the index after concatenation.

```python
import pandas as pd
# Dataset 1: Product details
df1 = pd.DataFrame({
    'ProductID': [101,102,103],
    'ProductName': ['Pen', 'Notebook', 'Marker'],
    'Price': [10, 50, 30]})


# Dataset 2: More products (same columns as df1 → good for vertical stacking)

df2 = pd.DataFrame({
    'ProductID': [104,105,106],
    'ProductName': ['Pencil', 'Eraser', 'Ruler'],
    'Price': [5, 20, 15]})

# Dataset 3: Additional info (different columns → good for horizontal stacking)

df3 = pd.DataFrame({
    'Stock': [100, 50, 75],
    'Rating': [4.5, 4.7, 4.8]
})

print(f"DF1 {df1}")
print(f"DF2 {df2}")
print(f"DF3 {df3}")
```

```python
#1. Vertical stacking (axis = 0)

# By default concat does vertical stacking

vertical_stacking = pd.concat([df1, df2])
print(f"Vertical Stacking:\n{vertical_stacking}")
```

```python
# To change index values

vertical_stacking = pd.concat([df1, df2], axis=0, ignore_index=True)
print(f"Vertical Stacking:\n{vertical_stacking}")
```

```python
#2. Horizontal Stacking (axis = 1)

horizontal_stacking = pd.concat([df1, df3], axis=1)
print(f"Horizontal Stacking:\n{horizontal_stacking}")
```

## 2. Using `merge()` for SQL-Style Joins
- `merge()` is similar to SQL joins.
- Specify `on` or `left_on`/`right_on` to choose key columns.
- Join types: `inner` (default), `left`, `right`, `outer`.

```python
# We have to concat df1 and df2 rows wise

Product = pd.concat([df1, df2], ignore_index=True)
print(f"Produce Df:\n{Product}")

Sales = pd.DataFrame({
    'ProductID': [101, 103, 105],
    'Sales': [1000, 800, 1200]
})

print(f"Sales Df:\n{Sales}")

Revenue = pd.DataFrame({
    'ProductID': [101, 102, 103],
    'Revenue': [2000, 1500, 2500]})

print(f"Revenue Df:\n{Revenue}")
```

```python
#1. Inner Join (Done Dataframes ke beech only matching ProductId)

pd.merge(Product,Sales, how='inner')
```

```python
#2. Left Join (Take left full table and only common element of Right Table)

pd.merge(Product,Sales, how='left')
```

```python
#3. Right Join (Take Right Table full and Left Table only common elements)

pd.merge(Product, Sales, how='right')
```

```python
#4. Outer join (Everthing, even unmatched)

outer = pd.merge(Product, Sales, how='outer')
print(f"Outer Join:\n{outer}")
```

## 4. Practical Examples
- Combine sales and revenue data to see performance.
- Merge students with marks to link records.
- Practice with realistic datasets to understand alignment and missing data handling.

```python
import pandas as pd

# Re-defining df1 and df2 as they are needed for Product DataFrame
df1 = pd.DataFrame({
    'ProductID': [101,102,103],
    'ProductName': ['Pen', 'Notebook', 'Marker'],
    'Price': [10, 50, 30]})

df2 = pd.DataFrame({
    'ProductID': [104,105,106],
    'ProductName': ['Pencil', 'Eraser', 'Ruler'],
    'Price': [5, 20, 15]})

# Re-defining Product, Sales, and Revenue DataFrames with 'ProductID'
Product = pd.concat([df1, df2], ignore_index=True)

Sales = pd.DataFrame({
    'ProductID': [101, 103, 105],
    'Sales': [1000, 800, 1200]
})

Revenue = pd.DataFrame({
    'ProductID': [101, 102, 103],
    'Revenue': [2000, 1500, 2500]})


print('--- 1. Combining Sales and Revenue Data ---')

# Merge Product with Sales data
product_sales = pd.merge(Product, Sales, on='ProductID', how='left')
print(f"Product and Sales Merged (Left Join):\n{product_sales}\n")

# Merge the combined data with Revenue data
combined_performance = pd.merge(product_sales, Revenue, on='ProductID', how='left')
print(f"Combined Performance (Product, Sales, Revenue - Left Join):\n{combined_performance}\n")

# Handle missing values (e.g., fill NaN with 0 for Sales and Revenue)
combined_performance_filled = combined_performance.fillna(0)
print(f"Combined Performance (NaN filled with 0):\n{combined_performance_filled}\n")

print('\n--- 2. Merging Students with Marks ---')

# Create sample Student DataFrame
students_df = pd.DataFrame({
    'StudentID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
})

# Create sample Marks DataFrame
marks_df = pd.DataFrame({
    'StudentID': [1, 2, 3, 5],
    'Subject': ['Math', 'Science', 'History', 'Math'],
    'Marks': [85, 92, 78, 65]
})

print(f"Students DataFrame:\n{students_df}\n")
print(f"Marks DataFrame:\n{marks_df}\n")

# Merge students with their marks (inner join to get only students with marks)
student_marks_inner = pd.merge(students_df, marks_df, on='StudentID', how='inner')
print(f"Student Marks (Inner Join):\n{student_marks_inner}\n")

# Merge students with their marks (left join to keep all students)
student_marks_left = pd.merge(students_df, marks_df, on='StudentID', how='left')
print(f"Student Marks (Left Join):\n{student_marks_left}\n")

# Analyze missing data in the left join
print("Missing data after left join:\n", student_marks_left.isnull().sum())
```

## Mini Practice
1. Create three DataFrames: Products, Sales, and Revenue.
2. Concatenate them vertically and horizontally.
3. Merge Sales and Revenue using different join types.
4. Join Products with Sales based on index or a key.
5. Analyze missing values after joins.

```python

# 1. Create three DataFrames: Products, Sales, and Revenue

import pandas as pd

Products = pd.DataFrame({
    'SKU': [101,102,103],
    'ProductName': ['Pen', 'Notebook', 'Marker'],
    'Price': [10, 50, 30]})


Sales = pd.DataFrame({
    'SKU': [101, 103, 105],
    'UnitsSold': [1000, 800, 1200]
})

Revenue = pd.DataFrame({
    'SKU': [101, 102, 103],
    'Revenue': [2000, 1500, 2500]})

print(f"Products:\n{Products}")
print(f"Sales:\n{Sales}")
print(f"Revenue:\n{Revenue}")


# 2. Concatenate them vertically and horizontally

concat_vertical = pd.concat([Products, Sales], axis=0, ignore_index=True)
print(f"Vertical Concat:\n{concat_vertical}")

concat_horizontal = pd.concat([Products, Revenue], axis=1)
print(f"Horizontal Concat:\n{concat_horizontal}")

# 3. Merge Sales and Revenue using different join types (inner, left, right, outer)

pd.merge(Sales, Revenue, how='inner')
pd.merge(Sales, Revenue, how='left')
pd.merge(Sales, Revenue, how='right')
pd.merge(Sales, Revenue, how='outer')

# 4. Join Products with Sales based on index or a key

p_indexed = Products.set_index('SKU')
joined = p_indexed.join(Sales)
print(f"Joined:\n{joined}")

# 5. Analyze missing values after joins
# Use isna(), info(), or describe() to explore gaps

print(joined.isna())
print(joined.info())
print(joined.describe())
```
