Welcome to **Class 8** of our Pandas series. Today we’ll learn how to handle and analyze time series data.

## 1. Converting Columns to Datetime (`pd.to_datetime`)
- Use `pd.to_datetime()` to convert strings or numbers into datetime objects.
- Once converted, Pandas recognizes them as dates, enabling powerful operations.

```python
import pandas as pd

# 1. Create the dataset
data = {
    "OrderID": [1001, 1002, 1003, 1004, 1005],
    "OrderDate": ["2024-01-05", "2024/01/12", "15-01-2024", "2024-01-20", "20240125"],
    "Customer": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Amount": [250, 400, 150, 300, 500]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors = "coerce")
df
```

```python
df['OrderDate'].dtype
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'], format = 'mixed', errors = 'coerce')
df
```

```python
pd.to_datetime(df['OrderDate'], errors = 'coerce', format = 'mixed').dt.strftime('%d-%m-%Y')
```

## 2. Setting DateTime Index & Resampling
- Set a datetime column as the index with `set_index()`.
- Resample time series using `resample('ME').mean()` or other rules ('D', 'W', 'YE').
- Useful for aggregating data by time periods.

```python
#Step 1 :- OrderDate set as Index

df_idex = df.set_index('OrderDate')
df_idex
```

```python
# Step 2 :- We will resample complete data
# according to month and we will get the total
# amount of that month

monthly_amount = df_idex.resample('ME')['Amount'].sum()
monthly_amount
```

```python
daily_amount = df_idex.resample('D')['Amount'].sum()
daily_amount
```

```python
yearly_avg = df_idex.resample('YE')['Amount'].mean()
yearly_avg
```

```python
weekly_avg = df_idex.resample('W')['Amount'].mean()
weekly_avg
```

## mini-project

Converting columns to datetime (pd.to_datetime)

Setting a DateTime index & resampling

```python

# 1️⃣ Create a dataset of website visits (string dates + visit counts)

df = pd.DataFrame({
    'VisitDate': ['2023-01-01', '2023-01-02',
                  '2023-01-03', '2023-01-04', '2023-01-05'],

    'VisitCounts': [100, 200,
                    300, 400, 500]
})

print(f"This is the original dataframe:\n {df.dtypes}")

# 2️⃣ Convert VisitDate to datetime

df['VisitDate'] = pd.to_datetime(df['VisitDate'], format = 'mixed', errors = 'coerce')

print(f"Converted VisitDate into Datetime:\n {df}")
print(df.dtypes)

# 3️⃣ Set VisitDate as the index

df = df.set_index('VisitDate')
print(f"Set VisitDate as the index:\n {df}")

# 4️⃣ Resample to get total visitors per month

monthly_visitors = df.resample('ME')['VisitCounts'].sum()
print(f"Total visitors per month:\n {monthly_visitors}")

# 5️⃣ Resample to get average visitors per week

weekly_avg_visitors = df.resample('W')['VisitCounts'].mean()
print(f"Average visitors per week:\n {weekly_avg_visitors}")
```
