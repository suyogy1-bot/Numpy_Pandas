# Q1. Fancy Indexing (1D Array)

## Problem Statement
Create the array:

```python
[10, 20, 30, 40, 50, 60]
```

Using fancy indexing, print:
- 10
- 30
- 60

## Sample Output
```python
[10 30 60]
```


````python
import numpy as np

arr = np.array([10,20,30,40,50,60])
result = arr[[0, 2, -1]]
print(result)
````


# Q3. Fancy Indexing (2D Array)

## Problem Statement
Create the matrix:

```python
[[10,20,30],
 [40,50,60],
 [70,80,90]]
```

Using fancy indexing, print:
- 10
- 50
- 90

## Sample Output
```python
[10 50 90]
```


````python
arr = np.array([[10,20,30],
                [40,50,60],
                [70,80,90]])

result = arr[[0,1,2],[0,1,2]]
print(result)
````


# Q4. Conditional Selection

## Problem Statement
Create the array:

```python
[5, 12, 18, 25, 30, 7]
```

Print only values greater than 15.

## Sample Output
```python
[18 25 30]
```


````python
arr = np.array([5, 12, 18, 25, 30, 7])

result = arr[arr > 15]
print(result)
````


# Q5. Conditional Selection

## Problem Statement
Create the array:

```python
[2, 4, 7, 10, 13, 16]
```

Print only even numbers.

## Sample Output
```python
[ 2  4 10 16]
```


````python
arr = np.array([2, 4, 7, 10, 13, 16])

result = arr[arr % 2 == 0]
print(result)
````


# Q7. Find Sorted Indices


# Q8. Concatenate Arrays

## Problem Statement
Create two arrays:

```python
arr1 = [1,2,3]
arr2 = [4,5,6]
```

Concatenate both arrays.

## Sample Output
```python
[1 2 3 4 5 6]
```


````python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

result = np.concatenate([arr1, arr2])
print(result)
````


# Q9. Vertical Stacking

## Problem Statement
Create two arrays:

```python
[1,2,3]
[4,5,6]
```

Combine them using vertical stacking.

## Sample Output
```python
[[1 2 3]
 [4 5 6]]
```


````python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

print(np.vstack([arr1,arr2]))
````


# Q10. Horizontal Stacking

## Problem Statement
Create two arrays:

```python
[10,20,30]
[40,50,60]
```

Combine them using horizontal stacking.

## Sample Output
```python
[10 20 30 40 50 60]
```


````python
arr1 = np.array([10,20,30])
arr2 = np.array([40,50,60])

print(np.hstack([arr1,arr2]))
````
