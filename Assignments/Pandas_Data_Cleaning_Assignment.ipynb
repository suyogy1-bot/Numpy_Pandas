## Q13. Mini Data Cleaning Task

Input:
data = {
    "Name":["Aman","Raj","Aman",None],
    "Age":[20,None,20,25],
    "City":["Bhopal","Delhi","Bhopal","Indore"]
}

Task:
1. Fill missing Age with mean.
2. Remove duplicate rows.
3. Display final DataFrame.

```python
data = { "Name":["Aman","Raj","Aman",None], 
        "Age":[20,None,20,25], 
        "City":["Bhopal","Delhi","Bhopal","Indore"]}

df = pd.DataFrame(data)
```

```python
a = df['Age'].mean()
a
```

```python
df['Age'] = df['Age'].fillna(a)
df
```

```python
df.drop_duplicates()
```
