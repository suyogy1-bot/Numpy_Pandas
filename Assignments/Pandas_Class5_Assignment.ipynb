## Q1. Sort Age in Ascending Order

Input:
Age = [25, 18, 30, 22]

Output:
18
22
25
30

```python
import pandas as pd
import numpy as np

Age = [25, 18, 30, 22]

age_series = pd.Series(Age)
sorted_age = age_series.sort_values()
sorted_age
```

## Q2. Sort Marks in Descending Order

## Q9. Find Percentage Frequency

Input:
City = ['A','A','B','C']

Output:
A = 50%
B = 25%
C = 25%

```python
City = pd.Series(['A','A',"B",'C'])

freq_percent = City.value_counts(normalize=True) * 100
freq_percent
```

## Q14. Student Statistics

Input:
Marks = [60,70,80,90]

Output:
Mean = 75
Median = 75
Max = 90
Min = 60

```python
Marks = pd.Series([60, 70, 80, 90])

print(Marks.mean(), Marks.median(), Marks.max(), Marks.min())
```
