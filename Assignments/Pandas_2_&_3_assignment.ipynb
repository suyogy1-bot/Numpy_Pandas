## Problem Statement
Create a DataFrame with Name, Age and City.

Using `loc`, display the Name and City of the first row.

```python
Name    Aman

```


# Q7. Rename a Column

## Problem Statement
Create a DataFrame:

```python
Name
Marks
```

Rename column `Marks` to `Score`.

## Sample Output
```python
Name  Score
```


# Q8. Handle Missing Values

## Problem Statement
Create a DataFrame:

```python
Name = ["Aman", None, "Neha"]
Marks = [80, 75, None]
```

Fill missing values using `fillna()`.

## Sample Output
```python
Missing values replaced
```


````python
df = pd.DataFrame({
    "Name" : ["Aman", None, "Neha"],
    "Marks" : [80, 75, None]
})

df_filled = df.fillna({
    "Name": "Unknown",
    "Marks": 0
})

print(df_filled)
````


# Q9. Replace Values

## Problem Statement
Create a DataFrame:

```python
City = ["BPL", "IND", "BPL"]
```

Replace:
- BPL → Bhopal
- IND → Indore

## Sample Output
```python
Bhopal
Indore
Bhopal
```


````python
df = pd.DataFrame({
    "City": ["BPL", "IND", "BPL"]})
df.replace(["BPL", "IND", "BPL"],["Bhopal","Indore","Bhopal"])
````
