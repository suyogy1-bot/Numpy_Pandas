## Q1. Create a 3×3 NumPy Matrix

### Input
```
1 2 3
4 5 6
7 8 9
```

### Output
```python
[[1 2 3]
 [4 5 6]
 [7 8 9]]
```


````python
import numpy as np
import pandas as pd

arr = np.matrix([[1,2,3],[4,5,6],[7,8,9]])
arr
````

## Q6. Create DataFrame
