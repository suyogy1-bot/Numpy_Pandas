# Q1. Display Unique Values

## Problem Statement
Create a DataFrame with a City column and display all unique cities.

## Sample Input
```python
City = ["Bhopal", "Indore", "Bhopal", "Delhi"]
```

## Sample Output
```python
['Bhopal' 'Indore' 'Delhi']
```


````python
import pandas as pd

City = pd.Series(["Bhopal", "Indore", "Bhopal", "Delhi"])

City.drop_duplicates()
````


# Q4. Group By City

## Problem Statement
Find average marks city-wise using groupby().

## Sample Input
```python
City   Marks
Bhopal 80
Bhopal 90
Delhi  70
```

## Sample Output
```python
Bhopal 85
Delhi  70
```


````python
df = pd.DataFrame({
    "City": ['Bhopal','Bhopal','Delhi'],
    "Marks": [80,90,70]
})
````

````python
a = df.groupby(by='City')
a.mean()
````


# Q8. Multiple Conditions

## Problem Statement
Display students whose marks are greater than 70 and age is greater than 20.

## Sample Output
```python
Filtered rows
```


````python
data = {
    "Student": ["Aman", "Riya", "Sahil", "Neha"],
    "Marks": [85, 72, 90,78],
    "Age": [20,25,30,40]
}

df = pd.DataFrame(data)
df
````


# Q9. Group By Department

## Problem Statement
Create Employee and Department data and find total salary department-wise.

## Sample Output
```python
HR       50000
IT      120000
```


````python
import pandas as pd

data = {
    "Employee": ["Aman", "Riya", "Sahil", "Neha", "John"],
    "Department": ["HR", "IT", "IT", "HR", "Finance"],
    "Salary": [30000, 60000, 60000, 20000, 50000],
    "City": ["Bhopal", "Delhi", "Indore", "Bhopal", "Mumbai"]
}

df = pd.DataFrame(data)

print(df)
````

````python
df.groupby('Department')['Salary'].sum()
````


# Q10. Complete Practice Question

## Problem Statement
Create a DataFrame containing:
- Name
- Age
- City
- Marks

Perform:
1. Show unique cities
2. Count city frequency
3. Filter marks > 80
4. Sort by marks

## Sample Output
```python
Required Results
```


````python
import pandas as pd

data = {
    "Name": ["Aman", "Riya", "Sahil", "Neha", "Suyog"],
    "Age":  [20, 22, 21, 23, 24],
    "City": ["Bhopal", "Indore", "Delhi", "Bhopal", "Mumbai"],
    "Marks": [75, 88, 92, 67, 81]
}

df = pd.DataFrame(data)

df
````

````python
df['City'].drop_duplicates()
````

````python
df['City'].value_counts()
````

````python
df[df['Marks'] > 80]
````

````python
df.sort_values(by='Marks')
````
